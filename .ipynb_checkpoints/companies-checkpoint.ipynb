```python
num = 2
preResponseTxt = ''
hrefNumList = []
while 1:
    baseUrl = 'https://www.chyxx.com/minglu/320115/' + 'list_' + str(num) + '.html'
    response = requests.get(baseUrl)
    if preResponseTxt == response.text:
        break
    hrefRaw = re.findall('<h3>.*?</h3>', response.text)
    
    for i in hrefRaw:
        href = re.findall('href=".*?"', i)
        hrefNum = re.findall('\d+', str(href))
        # print(hrefNum) 
        hrefNumList.append(hrefNum[0])
    
    preResponseTxt = response.text
    num += 1
    print(num)
hrefNumList
```

```python
hrefList = []
for i in hrefNumList:
    href = 'https://www.chyxx.com/minglu/' + i + '.html'
    # print(href)
    hrefList.append(href)
```

```python
companies = []
for i in hrefList:
    response = requests.get(i)
    raw = re.findall('<h1.*?</p>', response.text, re.S)
    fine1 = re.findall('>.*?<', raw[0])
    fine2 = []
    for i in fine1:
        m = i.replace('>', '')
        n = m.replace('<', '')
        fine2.append(n)
    fine3 = []
    for i in fine2:
        if i != '' and i != ' &gt; ':
            fine3.append(i)
    # print(len(fine3))
    fine4 = []
    if len(fine3) == 20:
        for i in [0, 7, 9, 11, 13, 15, 17, 19]:
            fine4.append(fine3[i])
        addr = fine3[3] + fine3[4] + fine3[5]
        fine4.append(addr) 
        companies.append(fine4)
    else:
        print(len(fine3))
        print(fine3)
    
companies
```
